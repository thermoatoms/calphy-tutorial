```python
from ase.io import read, write
import numpy as np

def random_substitute_elements(input_file, output_file, from_symbol='Cu', to_symbol='Ni', fraction=0.1, seed=None):

    # Load the structure
    atoms = read(input_file, format='lammps-data')

    # Get array of chemical symbols
    symbols = np.array(atoms.get_chemical_symbols())

    # Find indices of atoms to replace
    indices = np.where(symbols == from_symbol)[0]
    if len(indices) == 0:
        raise ValueError(f"No atoms of type '{from_symbol}' found in {input_file}")

    # Determine number to replace
    n_replace = int(len(indices) * fraction)
    if n_replace == 0:
        raise ValueError(f"Fraction too small, results in zero replacements.")

    print(f"Replacing {n_replace} of {len(indices)} '{from_symbol}' atoms with '{to_symbol}'")

    # Random selection (reproducible)
    rng = np.random.default_rng(seed)
    replace_indices = rng.choice(indices, size=n_replace, replace=False)

    # Perform replacement
    symbols[replace_indices] = to_symbol
    print(symbols)
    atoms.set_chemical_symbols(symbols)

    # Write out modified structure
    write(output_file, atoms, format='lammps-data')
    print(f"Modified structure written to {output_file}")


# Example usage:
# random_substitute_elements("pureCu.data", "CuNi_10.data", from_symbol='Cu', to_symbol='Ni', fraction=0.1, seed=42)
```

```python
random_substitute_elements('CuNi.lqd.data', 'Cu025Ni075.lqd.data', fraction=0.75)
```
